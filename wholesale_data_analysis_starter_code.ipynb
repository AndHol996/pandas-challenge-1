## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
subtotal=dataset['unit_price']*dataset['qty']
dataset["subtotal"]= subtotal
dataset.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def shipping(unit_weight,qty):
    total_weight=unit_weight*qty
    if total_weight > 50:
        shipping_price=total_weight*7
        return shipping_price
    elif total_weight <= 50:
        shipping_price=total_weight*10
        return shipping_price
#Sourced from Nelewout on https://stackoverflow.com/questions/34279378/apply-function-with-two-arguments-to-columns
dataset['shipping_price'] =dataset.apply(lambda x:shipping(x['unit_weight'], x['qty']), axis=1)
dataset.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
# Assuming sales tax includes shipping price--> sales tax=(subtotal+shipping price)* sales tax (%)/100% 
sales_tax=(dataset['subtotal']+dataset['shipping_price'])*(9.25/100)
total_price_raw=dataset['subtotal']+dataset['shipping_price']+sales_tax
total_price=round(total_price_raw,2)
dataset['total_price']=total_price
dataset.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
total_cost=dataset['qty']*dataset['unit_cost']+dataset['shipping_price']
dataset['total_cost']=total_cost
dataset.head()
```

```python
# Create a column for the profit of each line using line cost and line price
profit=dataset['total_price']-dataset['total_cost']
dataset['profit']=profit
dataset.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
client_list
total_revenue=[]
for i in client_list:
    revenue=dataset.loc[(dataset["client_id"]==i)]
    revenue_round=round(revenue['total_price'].sum(), 2)
   
    total_revenue.append(revenue_round)
   
    print(f"""Client ID: {i}
Total Revenue: ${revenue_round:,} \n""")

#print(total_revenue)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
client_list

total_revenue=[]
total_quantity=[]
total_shipping_price=[]
total_profit=[]
total_cost=[]

new_format={"client_id":client_list,
        'qty':total_quantity,
        'total_price':total_revenue,
        'shipping_price':total_shipping_price,
        'total_cost':total_cost,
        'profit':total_profit}

for i in client_list:
    
    handle=dataset.loc[(dataset["client_id"]==i)]

    quantity_round=round(handle['qty'].sum(),2)
    total_quantity.append(quantity_round)
    
    revenue_round=round(handle['total_price'].sum(),2)
    total_revenue.append(revenue_round)
   
    shipping_round=round(handle['shipping_price'].sum(),2)
    total_shipping_price.append(shipping_round)

    cost_round=round(handle['total_cost'].sum(),2)
    total_cost.append(cost_round)

    profit_round=round(handle['profit'].sum(),2)
    total_profit.append(profit_round)

new_df=pd.DataFrame(new_format)
new_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Rename the columns to reflect the change in the money format. 

renamed_df=new_df.rename(columns={"client_id": "Client ID", 
                       'qty':'Units',
                       'total_price':'Total Revenue ($ Millions)', 
                       'shipping_price':'Shipping ($ Millions)',
                       'profit':'Total Profit ($ Millions)',
                       'total_cost': 'Total Cost ($ Millions)'
                       })

reformated_df=renamed_df[['Client ID', 'Units',
                          'Shipping ($ Millions)', 'Total Revenue ($ Millions)',
                          'Total Cost ($ Millions)',
                          'Total Profit ($ Millions)']]

# Define the money columns.Influenced by https://www.datacamp.com/tutorial/python-select-columns 
money_col=reformated_df.loc[:,['Shipping ($ Millions)', 'Total Revenue ($ Millions)',
                          'Total Cost ($ Millions)','Total Profit ($ Millions)']]

# Define a function that converts a dollar amount to millions.
def currency_format_millions(row):
    return row/1000000
# Apply the currency_format_millions function to only the money columns. 
#Influenced by https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.update.html#pandas.DataFrame.update
df=money_col.apply(currency_format_millions, axis=1)

reformated_df.update(df)
reformated_df.head()
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
rdf=reformated_df.sort_values("Total Profit ($ Millions)", ascending=False)
rdf
```
